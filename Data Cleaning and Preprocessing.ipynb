```python
# Group the data by building type and calculate the mean for eui, wui, and ghg
energy_by_bldgtype = energy.groupby('bldgtype')[['eui', 'wui', 'ghg']].mean()
energy_by_bldgtype
```

```python
energy['zipcode'] = energy['zipcode'].astype(str).str[:5]
```

```python
energy_by_zipcode = energy.groupby(['zipcode'])[['eui', 'wui', 'ghg']].mean()
energy_by_zipcode = energy_by_zipcode.iloc[:-2]
energy_by_zipcode.dropna(inplace=True)
```

```python
energy_by_zipcode
```

```python
energy_by_zipcode_bldgtype = energy.groupby(['bldgtype', 'zipcode'])[['eui', 'wui', 'ghg']].mean()
energy_by_zipcode_bldgtype.dropna(inplace=True)
energy_by_zipcode_bldgtype.head(30)
```
